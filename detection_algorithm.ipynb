```python
class CoronaDetectionAlgorithm:
    def __init__(self):
        self.symptom_weights = {
            'temperature': 2,  # > 37.5°C
            'cough': 1,
            'fever': 2,
            'breathing_difficulty': 3,
            'loss_of_taste_smell': 2,
            'fatigue': 1,
            'chest_pain': 3
        }
        
        self.risk_factor_weights = {
            'travel_history': 2,
            'contact_with_infected': 3
        }
    
    def calculate_risk_score(self, symptoms, risk_factors):
        """Calculate corona risk score"""
        risk_score = 0
        
        # Temperature scoring
        if symptoms['temperature'] > 37.5:
            risk_score += self.symptom_weights['temperature']
        
        # Symptoms scoring
        for symptom, weight in self.symptom_weights.items():
            if symptom != 'temperature' and symptoms.get(symptom):
                risk_score += weight
        
        # Risk factors scoring
        for factor, weight in self.risk_factor_weights.items():
            if risk_factors.get(factor):
                risk_score += weight
        
        return risk_score
    
    def determine_result(self, risk_score, symptoms_count):
        """Determine test result based on risk score"""
        if risk_score >= 8:
            result = "Positive"
            if risk_score >= 12:
                severity = "High"
            elif risk_score >= 8:
                severity = "Medium"
            else:
                severity = "Low"
        else:
            result = "Negative"
            severity = "Low"
        
        return result, severity
    
    def get_recommendations(self, result, severity):
        """Get recommendations based on test result"""
        recommendations = {
            "Positive": {
                "High": [
                    "Immediate medical attention required",
                    "Isolate immediately",
                    "Monitor symptoms closely",
                    "Contact healthcare provider",
                    "Emergency services if severe breathing difficulty"
                ],
                "Medium": [
                    "Self-isolate for 14 days",
                    "Monitor symptoms daily",
                    "Consult healthcare provider",
                    "Rest and hydrate",
                    "Seek medical help if symptoms worsen"
                ],
                "Low": [
                    "Self-monitor for symptoms",
                    "Practice social distancing",
                    "Wear mask in public",
                    "Get retested if symptoms develop"
                ]
            },
            "Negative": {
                "Low": [
                    "Continue following safety guidelines",
                    "Monitor for symptoms",
                    "Practice good hygiene",
                    "Maintain social distancing"
                ]
            }
        }
        
        return recommendations.get(result, {}).get(severity, [])
```
